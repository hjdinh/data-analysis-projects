# Clean and Analyze Employee Exit Surveys

## Introduction

In this project, we'll work with exit surveys from employees of the Department of Education, Training and Employment (DETE) and the Technical and Further Education (TAFE) institute in Queensland, Australia. Slight modifications were made to these datasets to make them easier to work with, including changing the encoding to UTF-8 (the original ones are encoded using cp1252.)

In this project, we'll play the role of data analyst and pretend our stakeholders want to know the following:

- Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been there longer?

- Are younger employees resigning due to some kind of dissatisfaction? What about older employees?

They want us to combine the results for both surveys to answer these questions. However, although both used the same survey template, one of them customized some of the answers. The aim is to do most of the data cleaning and get started analyzing the first question.

```python
# import libraries
import pandas as pd
import numpy as np

# read in data
dete_survey = pd.read_csv('dete_survey.csv')
tafe_survey = pd.read_csv('tafe_survey.csv')

# explore data
dete_survey.info()
dete_survey.head()
dete_survey.isnull()
```

## Identify Missing Values and Drop Unnecessary Columns

The following observations can be made from reading the above datasets:

- The dete_survey dataframe contains 'Not Stated' values that indicate values are missing, but they aren't represented as NaN.

- Both the dete_survey and tafe_survey dataframes contain many columns that we don't need to complete our analysis.

- Each dataframe contains many of the same columns, but the column names are different.

- There are multiple columns/answers that indicate an employee resigned because they were dissatisfied.

First, we will specify the 'Not Stated' values as 'NaN'. Next, we will drop some columns from each dataframe that we won't use in our analysis.

```python
# read in data again but replace 'Not Stated' values with 'NaN'
dete_survey = pd.read_csv('dete_survey.csv', na_values='Not Stated')
```

```python
# remove columns not needed for analysis
dete_survey_updated = dete_survey.drop(dete_survey.columns[28:49], axis=1)
tafe_survey_updated = tafe_survey.drop(tafe_survey.columns[17:66], axis=1)

# check columns were dropped
print(dete_survey_updated.columns)
print(tafe_survey_updated.columns)
```

## Clean Column Names

Each dataframe contains many of the same columns, but the column names are different. Because we eventually want to combine them, we'll have to standardize the column names. 

```python
# clean column names
mapping = {'Record ID': 'id', 'CESSATION YEAR': 'cease_date', 'Reason for ceasing employment': 'separationtype', 'Gender. What is your Gender?': 'gender', 'CurrentAge. Current Age': 'age',
       'Employment Type. Employment Type': 'employment_status',
       'Classification. Classification': 'position',
       'LengthofServiceOverall. Overall Length of Service at Institute (in years)': 'institute_service',
       'LengthofServiceCurrent. Length of Service at current workplace (in years)': 'role_service'}

dete_survey_updated.columns = dete_survey_updated.columns.str.lower().str.strip().str.replace(' ','_')
tafe_survey_updated = tafe_survey_updated.rename(mapping, axis=1)

# check that columns were updated
dete_survey_updated.head()
```

```python
tafe_survey_updated.head()
```

## Filter the Data

If we look at the unique values in the separationtype columns in each dataframe, we'll see that each contains a couple of different separation types. For this project, we'll only analyze survey respondents who resigned, so their separation type contains the string 'Resignation'.

```python
# check unique values for separationtype column
dete_survey_updated['separationtype'].value_counts()
```

```python
# check unique values for separationtype column
tafe_survey_updated['separationtype'].value_counts()
```

```python
# Update all separation types containing the word "resignation" to 'Resignation'
dete_survey_updated['separationtype'] = dete_survey_updated['separationtype'].str.split('-').str[0]

# Check the values in the separationtype column were updated correctly
dete_survey_updated['separationtype'].value_counts()
```
